```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1=np.arange(10)
x_2=np.arange(1,21,2)
y=np.zeros(10,dtype=int)
y=y+(3*x_1+2*x_2+3)
print(x_1.size)
```

```python
plt.yticks(np.arange(0, max(y)+1, 2))
plt.plot(x_1,x_2,y,marker='o')
```
