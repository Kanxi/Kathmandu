```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y=np.zeros(10,dtype=int)
y=y+(2*x+3)
```

```python
plt.yticks(np.arange(0, max(y)+1, 2))
plt.plot(x,y,marker='o')
```
