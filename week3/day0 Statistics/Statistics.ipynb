```python
import matplotlib.pyplot as plt

# Right-skewed data
left_skewed_data = {
    5: 1,
    6: 2,
    7: 3,
    8: 4,
    9: 5,
    10: 6,
    11: 7,
    12: 8,
    13: 9,
    14: 10
}

# Left-skewed data
right_skewed_data = {
    14: 1,
    13: 2,
    12: 3,
    11: 4,
    10: 5,
    9: 6,
    8: 7,
    7: 8,
    6: 9,
    5: 10
}
# No-skew data
no_skew_data = [8, 10, 9, 11, 10, 9, 12, 11, 8, 10]

# Plotting histograms
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(right_skewed_data, bins=5, color='skyblue', edgecolor='black')
plt.title('Right-skewed Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(left_skewed_data, bins=5, color='salmon', edgecolor='black')
plt.title('Left-skewed Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(no_skew_data, bins=5, color='lightgreen', edgecolor='black')
plt.title('No-skew Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

# plt.tight_layout()
plt.show()
```
